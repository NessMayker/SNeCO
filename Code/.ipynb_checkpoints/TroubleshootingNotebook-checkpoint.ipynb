```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import fits

database = Table.read('../Data/3.SNe+GalData.csv', format = "csv")

database[0:1]
```

```python
map = database["MixedMaps"][35]
paMap = database["ImageFile150"][35]
starMap = database["IFRResizedMaps"][35]
```

```python
mixedMap = fits.open(map)
mapData = mixedMap[0].data

paMap = fits.open(paMap)
paData = paMap[0].data

starImg = fits.open(starMap)
stars = starImg[0].data
```

```python
f, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (15,5))
ax[0].imshow(mapData)
ax[1].imshow(paData)
ax[2].imshow(stars)
```
