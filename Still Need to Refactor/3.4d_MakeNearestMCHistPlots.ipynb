```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from astropy.table import Table
```

```python
database = Table.read('../Data/3.SNe+GalData.csv', format = "csv")
histTable = Table.read('../Data/3.HistData.csv', format='csv')
histTable[0:1]
```

```python
# Make histogram of Nearest MC 10^5.5 & 6.5

sns.set_theme(style = "whitegrid")

f, ax = plt.subplots(nrows = 1, ncols = 2, sharex = False, sharey = True, figsize = (8,4))

sns.color_palette("dark")

sns.histplot(histTable['MCSN55Int'])
```
